```python
# Filter PRS data and merge datasets for each dataset
def filter_and_merge(df_prs, df_cc):
    df_prs_filtered = df_prs[df_prs['In_Regression'] == 'Yes'].dropna(subset=['PRS'])
    df_prs_filtered['PRS'] = pd.to_numeric(df_prs_filtered['PRS'], errors='coerce')
    merged_df = pd.merge(df_prs_filtered, df_cc, on=['FID', 'IID'])
    return merged_df

```

```python


# Plot distribution curves for each dataset
def plot_distribution(df_prs, df_cc, title, ax):
    # Filter PRS data
    df_prs_filtered = df_prs[df_prs['In_Regression'] == 'Yes'].dropna(subset=['PRS'])
    df_prs_filtered['PRS'] = pd.to_numeric(df_prs_filtered['PRS'], errors='coerce')

    # Merge datasets
    merged_df = pd.merge(df_prs_filtered, df_cc, on=['FID', 'IID'])

    # Standardize the PRS data
    scaler = StandardScaler()
    merged_df['PRS_standardized'] = scaler.fit_transform(merged_df['PRS'].values.reshape(-1, 1))

    # Plot distribution curves for cases and controls
    sns.histplot(merged_df[merged_df['Phenotypes'] == 1]['PRS_standardized'], kde=True, color='orange', label='Cases', stat='density', ax=ax)
    sns.histplot(merged_df[merged_df['Phenotypes'] == 0]['PRS_standardized'], kde=True, color='dodgerblue', label='Controls', stat='density', ax=ax)

 
 # Set title and remove individual axis labels
    ax.set_title(title)
    ax.set_xlabel('')
    ax.set_ylabel('')


# Create subplots in two rows, three columns
fig, axes = plt.subplots(3, 3, figsize=(15, 10))

# Plot distribution curves for each dataset
plot_distribution(df_prs1, df_cc1, 'Craig all patients for POAG', axes[0, 0])
plot_distribution(df_prs2, df_cc2, 'Craig all patients for IOP', axes[0, 1])
plot_distribution(df_prs3, df_cc3, 'Craig all patients for VCDR', axes[0, 2])

# Plot distribution curves for IOP GWAS datasets
plot_distribution(df_prs4, df_cc4, 'Craig TD for POAG', axes[1, 0])
plot_distribution(df_prs5, df_cc5, 'Craig TD for IOP', axes[1, 1])
plot_distribution(df_prs6, df_cc6, 'Craig TD for VCDR', axes[1, 2])

plot_distribution(df_prs7, df_cc7, 'POAG clumped list for all patients', axes[2, 0])
plot_distribution(df_prs6, df_cc6, 'POAG clumped list for TD', axes[2, 1])


# Add a legend for all subplots
handles, labels = axes[0, 0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper right',bbox_to_anchor=(1.05, 1.05))

# Add common x and y axis labels
fig.text(0.5, -0.05, 'Standardized PRS', ha='center', fontsize=15)
fig.text(0.0001, 0.5, 'Density', va='center', rotation='vertical', fontsize=15, ha='left')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Plot deciles for each dataset
def plot_deciles(df_prs, df_cc, title, ax):
    # Filter PRS data
    df_prs_filtered = df_prs[df_prs['In_Regression'] == 'Yes'].dropna(subset=['PRS'])
    df_prs_filtered['PRS'] = pd.to_numeric(df_prs_filtered['PRS'], errors='coerce')

    # Merge datasets
    merged_df = pd.merge(df_prs_filtered, df_cc, on=['FID', 'IID'])

    # Calculate PRS deciles
    merged_df['PRS_decile'] = pd.qcut(merged_df['PRS'], q=10, labels=False)

    # Count the number of individuals in each decile for cases and controls
    decile_counts = merged_df.groupby(['PRS_decile', 'Phenotypes']).size().unstack(fill_value=0)

    # Calculate proportions
    decile_proportions = decile_counts.div(decile_counts.sum(axis=1), axis=0)

    # Plot decile proportions
    barplot = sns.barplot(x=decile_proportions.index, y=decile_proportions[1], color='darkorange', label='Cases', ax=ax)
    sns.barplot(x=decile_proportions.index, y=decile_proportions[0], color='skyblue', bottom=decile_proportions[1], ax=ax)

    # Annotate each bar with separate counts for cases
    for p, case_count in zip(barplot.patches, decile_counts[1]):
        height = p.get_height()
        ymin, ymax = ax.get_ylim()
        position = ymax - 0.99 * (ymax - ymin) 
        ax.text(p.get_x() + p.get_width() / 2,
                 position,
                 f'{case_count}',
                 ha='center')

    # Annotate each bar with separate counts for controls
    for p, control_count in zip(barplot.patches, decile_counts[0]):
        height = p.get_height() + decile_proportions[1]
        ymin, ymax = ax.get_ylim()
        position = ymax - 0.1 * (ymax - ymin) 
        ax.text(p.get_x() + p.get_width() / 2,
                 position,
                 f'{control_count}',
                 ha='center')
    
    # Set title and remove individual axis labels and legend
    ax.set_title(title)
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.get_legend().remove()

# Create subplots in a single row
fig, axes = plt.subplots(3, 3, figsize=(18, 12))


# Plot deciles for each dataset
plot_deciles(df_prs1, df_cc1, 'Craig all patients for POAG', axes[0, 0])
plot_deciles(df_prs2, df_cc2, 'Craig all patients for IOP', axes[0, 1])
plot_deciles(df_prs3, df_cc3, 'Craig all patients for VCDR', axes[0, 2])

plot_deciles(df_prs4, df_cc4, 'Craig TD for POAG', axes[1, 0])
plot_deciles(df_prs5, df_cc5, 'Craig TD for IOP', axes[1, 1])
plot_deciles(df_prs6, df_cc6, 'Craig TD for VCDR', axes[1, 2])

plot_deciles(df_prs7, df_cc7, 'POAG clumped list for all patients', axes[2, 0])
plot_deciles(df_prs6, df_cc6, 'POAG clumped list for TD', axes[2, 1])



# Define handles and labels for the legend
handles = [plt.Line2D([0,0],[0,0], color='darkorange', lw=4),
           plt.Line2D([0,0],[0,0], color='skyblue', lw=4)]
labels = ['Cases', 'Controls']

# Create an empty subplot for the legend
fig.add_subplot(111, frameon=False)
plt.tick_params(labelcolor='none', top=False, bottom=False, left=False, right=False)
plt.xlabel('PRS Decile', fontsize=15)
plt.ylabel('Proportion', fontsize=15)
plt.legend(handles, labels, loc='upper left', bbox_to_anchor=(1, 1))

# Adjust layout
plt.subplots_adjust(right=0.85)

# Show the plots
plt.show()
```

```python
from sklearn.utils import resample
import numpy as np

def calculate_auc_ci(y_true, y_scores, n_bootstraps=1000, ci_level=0.95):
    bootstrapped_scores = []
    rng = np.random.RandomState(42)

    for _ in range(n_bootstraps):
        indices = rng.randint(0, len(y_scores), len(y_scores))
        if len(np.unique(y_true[indices])) < 2:
            continue

        score = auc(roc_curve(y_true[indices], y_scores[indices])[0],
                    roc_curve(y_true[indices], y_scores[indices])[1])
        bootstrapped_scores.append(score)

    sorted_scores = np.array(bootstrapped_scores)
    sorted_scores.sort()

    ci_lower = sorted_scores[int((1.0 - ci_level) / 2.0 * len(sorted_scores))]
    ci_upper = sorted_scores[int((1.0 + ci_level) / 2.0 * len(sorted_scores))]

    return ci_lower, ci_upper

# Calculate ROC curve, AUC and CI for each dataset
def calculate_roc_auc_ci(df, phenotype_col, prs_col):
    y_true = df[phenotype_col].values
    y_scores = df[prs_col].values
    fpr, tpr, _ = roc_curve(y_true, y_scores)
    roc_auc = auc(fpr, tpr)
    ci_lower, ci_upper = calculate_auc_ci(y_true, y_scores)
    return fpr, tpr, roc_auc, ci_lower, ci_upper

fpr1, tpr1, roc_auc1, ci_lower1, ci_upper1 = calculate_roc_auc_ci(merged_df1, 'Phenotypes', 'PRS')
fpr2, tpr2, roc_auc2, ci_lower2, ci_upper2 = calculate_roc_auc_ci(merged_df2, 'Phenotypes', 'PRS')
fpr3, tpr3, roc_auc3, ci_lower3, ci_upper3 = calculate_roc_auc_ci(merged_df3, 'Phenotypes', 'PRS')
fpr4, tpr4, roc_auc4, ci_lower4, ci_upper4 = calculate_roc_auc_ci(merged_df4, 'Phenotypes', 'PRS')
fpr5, tpr5, roc_auc5, ci_lower5, ci_upper5 = calculate_roc_auc_ci(merged_df5, 'Phenotypes', 'PRS')
fpr6, tpr6, roc_auc6, ci_lower6, ci_upper6 = calculate_roc_auc_ci(merged_df6, 'Phenotypes', 'PRS')
fpr7, tpr7, roc_auc7, ci_lower7, ci_upper7 = calculate_roc_auc_ci(merged_df7, 'Phenotypes', 'PRS')
fpr8, tpr8, roc_auc8, ci_lower8, ci_upper8 = calculate_roc_auc_ci(merged_df8, 'Phenotypes', 'PRS')

# Print the AUCs and their CIs
print(f'POAG AUC: {roc_auc1:.2f} (95% CI: {ci_lower1:.2f} - {ci_upper1:.2f})')
print(f'IOP AUC: {roc_auc2:.2f} (95% CI: {ci_lower2:.2f} - {ci_upper2:.2f})')
print(f'VCDR AUC: {roc_auc3:.2f} (95% CI: {ci_lower3:.2f} - {ci_upper3:.2f})')
print(f'POAG (TD) AUC: {roc_auc4:.2f} (95% CI: {ci_lower4:.2f} - {ci_upper4:.2f})')
print(f'IOP (TD) AUC: {roc_auc5:.2f} (95% CI: {ci_lower5:.2f} - {ci_upper5:.2f})')
print(f'VCDR (TD) AUC: {roc_auc6:.2f} (95% CI: {ci_lower6:.2f} - {ci_upper6:.2f})')
print(f'POAG clumped AUC: {roc_auc7:.2f} (95% CI: {ci_lower7:.2f} - {ci_upper7:.2f})')
print(f'POAG clumped (TD) AUC: {roc_auc8:.2f} (95% CI: {ci_lower8:.2f} - {ci_upper8:.2f})')

```
